```python
from PIL import Image, ImageOps
from sklearn.cluster import KMeans
import numpy as np
```

```python
image = Image.open('image.jpg')
compressed = ImageOps.fit(image, (128, 128))
values = compressed.getdata()
```

```python
# Square the values, because humans perceive color on log-scale
sqvals = [list(map(lambda x: x**2, value)) for value in values]
```

```python
processed = np.array(sqvals)
kmeans = KMeans(n_clusters=7, random_state=0, verbose=True).fit(processed)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
def make_hashval(r, g, b):
    return '#%02x%02x%02x' % (int(r), int(g), int(b))
```

```python
def display_color(color):
    figure = plt.figure()
    ax1 = figure.add_subplot(111, aspect='equal')
    ax1.add_patch(
        patches.Rectangle(
            (0.1, 0.1),   # (x,y)
            0.5,          # width
            0.5,          # height
            facecolor=color
        )
    )
```

```python
import math
squared = kmeans.cluster_centers_
normalized = [list(map(lambda x: math.sqrt(x), val)) for val in squared]
colors = [make_hashval(rgb[0], rgb[1], rgb[2]) for rgb in normalized]
```

```python
for color in colors:
    display_color(color)
```
